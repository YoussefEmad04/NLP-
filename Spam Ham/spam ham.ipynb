```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sms=pd.read_csv('spam.csv',encoding='ISO-8859-1')
sms
```

```python
sms.head()
```

```python
sms.head()
```

```python
X = sms['v2']
X
```

```python
y=sms['v1'].map({'spam':1,'ham':0})
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer(num_words=20000)
```

```python
tokenizer.fit_on_texts(X_train)
```

```python
seq_train=tokenizer.texts_to_sequences(X_train)
seq_train
```

```python
seq_test=tokenizer.texts_to_sequences(X_test)
seq_test
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
data_train=pad_sequences(seq_train)
data_train
```

```python
data_test = pad_sequences(seq_test, maxlen=data_train.shape[1])
data_test
```

```python
data_train.shape
```

```python
data_test.shape
```

```python
from tensorflow.keras.layers import Dense, Input, GlobalMaxPooling1D
from tensorflow.keras.layers import LSTM, Embedding
from tensorflow.keras.models import Model
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Flatten, Dense

# Assuming you have max_words and max_len defined somewhere
max_words = 10000
max_len = 189

model = Sequential()

# embedding layer
model.add(Embedding(max_words, 50, input_length=max_len, trainable=True))

# lstm layer
model.add(LSTM(100, return_sequences=True, dropout=0.5))
model.add(Flatten())

# Dense layer
model.add(Dense(200, activation='relu', name='hl1'))  # hidden layer 1
model.add(Dense(100, activation='relu', name='hl2'))  # hidden layer 2
model.add(Dense(1, activation='sigmoid', name='ol'))  # output layer
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
hist = model.fit(x=data_train, y=y_train, epochs=20, validation_data=(data_test, y_test))
```

```python
plt.plot(hist.history['loss'], label='Loss')
plt.plot(hist.history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()
```

```python
plt.plot(hist.history['accuracy'], label='Accuracy')
plt.plot(hist.history['val_accuracy'], label='Validation accuracy')
plt.legend()
plt.show()
```
