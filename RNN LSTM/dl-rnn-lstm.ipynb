## Google Stock Data | RNN Basics Case-study

### Data Importing & Inspecting

```python
dataset_train = pd.read_csv('Google_Stock_Price_Train.csv')
dataset_train
```

```python
# open price
training_set = dataset_train.iloc[:, 1:2].values
training_set
```

### Preprocessing Data

##### Scaling Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler(feature_range = (0, 1))
```

```python
training_set_scaled = sc.fit_transform(training_set)
training_set_scaled
```

```python
training_set_scaled.shape
```

```python
# el wahed dah eli ana 3ayz a3mlo predicted
```

##### Selecting Data

```python
### Creating a data structure with 60 timesteps and 1 output
X_train = []
y_train = []
for i in range(60, len(training_set_scaled)):
    X_train.append(training_set_scaled[i-60:i, 0]) # i-60 =0 mn 0 L i historical data hselect haba w ha5leh twq3 eli b3dehom
    y_train.append(training_set_scaled[i, 0]) # eli b3d el qym eli 5dtha abl kda
```

```python
X_train
```

```python
y_train # el output eli b3d el 60 w kda lw hat7rk mn 3 L 7 yb2a eli b3deh ehh 8 w kda ba3ml data set llinput w datat set lloutput
```

```python
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
pd.DataFrame(X_train).head(5)
```

```python
X_train.shape # kolo 1198 w 60 input 
```

```python
pd.DataFrame(y_train).head(5)
```

##### Reshaping Data

```python
X_train.shape[0]
```

```python
X_train.shape[1]
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))# el whed dah el output ll60 eli kant fo2
```

```python
X_train.shape
```

```python
y_train.shape
```

### Model Building & Training 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

```python
regressor = Sequential()

regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
regressor.add(Dropout(0.2)) # dropout cause of regulaization to reduce overfitting

regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50)) # m4 lazem arg3 sequnces fe el a5r
regressor.add(Dropout(0.2))

regressor.add(Dense(units = 1))
```

```python
regressor.summary()
```

```python
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error') #34an dah regression ana bt3aml m3 arkam m4 classififcation 
```

```python
regressor.fit(X_train, y_train, epochs = 100, batch_size = 32)
```

### Model Evaluation

##### Preparing Data

```python
dataset_test = pd.read_csv('Google_Stock_Price_Test.csv')
dataset_test
```

```python
real_stock_price = dataset_test.iloc[:, 1:2].values
real_stock_price
```

```python
dataset_total = pd.concat((dataset_train['Open'], dataset_test['Open']), axis = 0)
dataset_total
```

```python
inputs = dataset_total[len(dataset_total) - len(dataset_test)  - 60:].values
inputs #data set length equal 20
```

```python
inputs.shape
```

```python
inputs = inputs.reshape(-1,1)
inputs = sc.transform(inputs)
```

```python
inputs.shape #leeh 80 w wahed 34an aa2dr astntg el you eli b3do
```

```python
inputs
```

```python
X_test = []
for i in range(60, 80):
    X_test.append(inputs[i-60:i, 0])
```

```python
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
X_test
```

```python
X_test.shape
```

##### Evaluating Model

```python
predicted_stock_price = regressor.predict(X_test)
predicted_stock_price
```

```python
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
predicted_stock_price #inverse transform 34an bd ma homa mn 0 L 1 la ana arg3hom laslohm
```

```python
plt.plot(real_stock_price, color = 'red', label = 'Real Google Stock Price')
plt.plot(predicted_stock_price, color = 'blue', label = 'Predicted Google Stock Price')
plt.title('Google Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Google Stock Price')
plt.legend()
plt.show()
```
